```python
# Number Addtion
import pandas as pd
from fontTools.varLib.avarPlanner import measureWeight


def cleanPrice(listOfPrice, itemToClean):
    cleanedPrice = []
    for i in range(len(listOfPrice)):
        listOfPrice[i] = listOfPrice[i].replace(itemToClean, "")
        cleanedPrice.append(listOfPrice[i])
    return cleanedPrice
```
