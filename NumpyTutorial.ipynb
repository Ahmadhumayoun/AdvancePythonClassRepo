```python
import timeit
import numpy as np

# Setup: 1 million elements
size = 1000000
setup = f"import numpy as np; my_list = list(range({size})); my_arr = np.arange({size})"

# List operation: Square each number (list comprehension)
list_time = timeit.timeit("[x**2 for x in my_list]", setup=setup, number=10)

# Array operation: Square each number (vectorized)
arr_time = timeit.timeit("my_arr**2", setup=setup, number=10)

print(f"List time: {list_time:.5f}s")
print(f"Array time: {arr_time:.5f}s")
print(f"Array is {list_time/arr_time:.2f}x faster")
```
