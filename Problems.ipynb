```python
ProductList = ["Biscuits","Chips","Meat","Eggs","Bread"]
PriceList = ["20rs","50rs","500rs","300rs","100rs"]
cart = []
totalPrice = 0
```

```python
while True:
    print("Enter 1 for Biscuits\n Enter 2 for Chips\n Enter 3 for Meat\n Enter 4 for Eggs\n Enter 5 for Bread")
    choice = input("")
    print(f"How much do you want to buy?")
    quantity = int(input(""))
    if choice == "1":
        for i in range(quantity):
            cart.append(ProductList[0])
            totalPrice = totalPrice + int(PriceList[0].replace("rs",""))
    elif choice == "2":
        for i in range(quantity):
            cart.append(ProductList[1])
            totalPrice = totalPrice + int(PriceList[1].replace("rs",""))
    elif choice == "3":
        for i in range(quantity):
            cart.append(ProductList[2])
            totalPrice = totalPrice + int(PriceList[2].replace("rs",""))
    elif  choice == "4":
        for i in range(quantity):
            cart.append(ProductList[3])
            totalPrice = totalPrice + int(PriceList[3].replace("rs",""))
    elif choice == "5":
        for i in range(quantity):
            cart.append(ProductList[4])
            totalPrice = totalPrice + int(PriceList[4].replace("rs",""))
    print("Do you want to shop more?y/n")
    answer = input("")
    if answer == "n":
        break
    else:
        continue
```

```python
print(cart)
print(totalPrice)
```
