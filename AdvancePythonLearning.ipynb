```python
class userProfile:
    def __init__(self,userName,userAge):
        self.userName = userName
        self.userAge = userAge


    def addLastName(self,lastName):
        self.userName = self.userName + " " + lastName
    def printUserName(self):
        print(self.userName)
    def printUserAge(self):
        print(f"{self.userName} is {20-18}")
    def printUserAddress(self):
        print(f"{self.userName} lives in New York")
```

```python
u = userProfile("Ahmad",25)
u.addLastName("humayoun")
u.printUserName()
u.printUserAge()
u.printUserAddress()
```

```python
class pclass:

    def parentClassFunction(self):
        print("this function is from parent class")

class cClass(pclass):

    def childClassFunction (self):
        print("this function is from child class")

Child = cClass()
Child.parentClassFunction()
Child.childClassFunction()
```

```python
class parentClass:
    def __init__(self,_name):
        self.__name = _name

    def publicFunc(self):
        print("this function is from parent class")
    def privateFunc(self):
        print("this function is from parent class")

class childClass(parentClass):
    def __init__(self,name,age):
        super().__init__(name)
        self.age = age

    def chilClassFunc(self):
        print("this function is from child class")
class grandChildClass(childClass):
    def __init__(self,name,age):
        super().__init__(name,age)
        self.age = age
```

```python
ca = grandChildClass("Ahmad",20)
```

```python

ca.privateFunc()
```

```python
class parentClass1:
    def polyFunction(self):
        print("this function is from parent class")
class childClass1(parentClass1):
    def polyFunction(self):
        print("this function is from child class")
```
