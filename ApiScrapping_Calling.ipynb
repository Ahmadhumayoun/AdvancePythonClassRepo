```python

url = 'https://api.escuelajs.co/api/v1/auth/login'
h = {
    'Content-Type': 'application/json'
}
accesToken = ""
credentials = {"email":"john@mail.com","password":"changeme"}
response = r.post(url,json = credentials,headers=h)
print(response.json())
postresponse = response.json()
accessToken = postresponse['access_token']
print(accessToken)
```
