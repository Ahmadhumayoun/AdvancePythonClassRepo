```python
print(" Unit Conversion System ")
print("1. Length")
print("2. Temperature")
print("3. Weight")
print("4. Time")

choice = int(input("Select a category (1-4): "))

# LENGTH
if choice == 1:
    print("\nLength Conversion")
    print("1. Meters to Kilometers")
    print("2. Kilometers to Meters")
    print("3. Inches to Feet")
    print("4. Feet to Inches")
    option = int(input("Choose conversion: "))
    value = float(input("Enter value: "))
    if option == 1:
        result = value / 1000
        print("Result =", result, "kilometers")
    elif option == 2:
        result = value * 1000
        print("Result =", result, "meters")
    elif option == 3:
        result = value / 12
        print("Result =", result, "feet")
    elif option == 4:
        result = value * 12
        print("Result =", result, "inches")
# TEMPERATURE
elif choice == 2:
    print("\nTemperature Conversion")
    print("1. Celsius to Fahrenheit")
    print("2. Fahrenheit to Celsius")
    option = int(input("Choose conversion: "))
    value = float(input("Enter value: "))
    if option == 1:
        result = (value * 9/5) + 32
        print("Result =", result, "F")
    elif option == 2:
        result = (value - 32) * 5/9
        print("Result =", result, "C")
# WEIGHT
elif choice == 3:
    print("\nWeight Conversion")
    print("1. Kilograms to Grams")
    print("2. Grams to Kilograms")
    print("3. Kilograms to Pounds")
    option = int(input("Choose conversion: "))
    value = float(input("Enter value: "))
    if option == 1:
        result = value * 1000
        print("Result =", result, "grams")
    elif option == 2:
        result = value / 1000
        print("Result =", result, "kg")
    elif option == 3:
        result = value * 2.205
        print("Result =", result, "pounds")
# TIME
elif choice == 4:
    print("\nTime Conversion")
    print("1. Seconds to Minutes")
    print("2. Minutes to Seconds")
    print("3. Minutes to Hours")
    print("4. Hours to Minutes")
    option = int(input("Choose conversion: "))
    value = float(input("Enter value: "))
    if option == 1:
        result = value / 60
        print("Result =", result, "minutes")
    elif option == 2:
        result = value * 60
        print("Result =", result, "seconds")
    elif option == 3:
        result = value / 60
        print("Result =", result, "hours")
    elif option == 4:
        result = value * 60
        print("Result =", result, "minutes")
# INVALID
else:
    print("Invalid choice!")
```
