```python
from tqdm import tqdm
import pandas as pd
import requests as r
from bs4 import BeautifulSoup as bs
```

```python
cardsUrl ="https://www.deckshop.pro/card/list"
response =r.get(cardsUrl)
```

```python
with open("CardsUrl.html","w") as file:
    file.write(response.text)
```

```python
baseUrl = "https://www.deckshop.pro"
with open ("CardsUrl.html","r") as file:
    sourceCode = file.read()
    soup = bs(sourceCode, "html.parser")
    cardsUrls = soup.find_all("img",{"class":"card"})
    cardDictioanry = {"Cards Name":[],"Cards Image Urls": []}
    for eachUrl in tqdm( cardsUrls,desc="Downloading Cards",unit="img"):
        cardDictioanry["Cards Name"].append( eachUrl.get("src").split("/")[-1].split(".")[0])
        cardDictioanry["Cards Image Urls"].append(eachUrl.get("src"))
        # tqdm.write(f"Downloading {eachUrl.get('src').split('/')[-1]}...")
        downloadedImage = r.get(baseUrl+eachUrl.get("src"))
        with open("CardsDownload/"+eachUrl.get("src").split("/")[-1],"wb") as img:

            img.write(downloadedImage.content)
    cardDataFrame = pd.DataFrame(cardDictioanry)
    cardDataFrame




```

```python
tableresponse = r.get("https://www.scrapethissite.com/pages/forms/")
with open("TableData.html","w") as file:
    file.write(tableresponse.text)
```

```python
with open("TableData.html","r") as file:
    sourceCode = file.read()
    soup = bs(sourceCode,"html.parser")
    tableData = soup.find_all("tr")
    tableDataList = []
    for eachRow in tableData:
        cell = eachRow.find_all(["td","th"])
        singleRow=[]
        for eachCell in cell:
            singleRow.append(eachCell.get_text().strip())
        tableDataList.append(singleRow)
```

```python
tableDF = pd.DataFrame(tableDataList[1:],columns=tableDataList[0])
```
