```python


# --- 1. Data Input ---
# The input() function always returns data as a string
user_name = input("Please enter your name: ")
user_location = input("Where are you joining us from? ")

# --- 2. String Concatenation & Formatting ---
# Method A: Basic Concatenation using +
welcome_basic = "Welcome, " + user_name + " from " + user_location + "!"

# Method B: f-strings (The modern, preferred way)
welcome_fancy = f"Identity Verified: Hello {user_name}. Access granted from {user_location}."

# --- 3. Output ---
print("-" * 30)
print(welcome_basic)
print(welcome_fancy)
print(f"You are now running on System Version {SYSTEM_VERSION}.")
print("-" * 30)
```
