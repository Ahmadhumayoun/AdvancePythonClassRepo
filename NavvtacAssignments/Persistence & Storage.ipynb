```python
import json

# --- 1. Reading the Raw TXT File ---
# Let's assume 'logs.txt' contains lines like:
# 2026-04-10,System_Boot,Success
# 2026-04-10,User_Login,Failure

log_data = []

try:
    # 'r' mode stands for Read
    with open("logs.txt", "r") as file:
        for line in file:
            # Clean the line and split by comma
            parts = line.strip().split(",")
            if len(parts) == 3:
                entry = {
                    "date": parts[0],
                    "event": parts[1],
                    "status": parts[2]
                }
                log_data.append(entry)
except FileNotFoundError:
    print("Error: 'logs.txt' not found. Creating dummy data for demonstration.")
    log_data = [{"date": "2026-04-10", "event": "Manual_Entry", "status": "Fixed"}]

# --- 2. Saving to JSON ---
# 'w' mode stands for Write (this overwrites existing content)
with open("formatted_logs.json", "w") as json_file:
    # indent=4 makes the JSON human-readable
    json.dump(log_data, json_file, indent=4)

print("Conversion Complete: 'logs.txt' -> 'formatted_logs.json'")
```
