```python
# --- 1. Basic List Operations ---
tasks = ["Email Client", "Update Server", "Urgent: Fix Bug", "Prepare Slides"]

# Adding a new task
tasks.append("Review Logs")

# Removing the last task (the one just added)
completed_task = tasks.pop()

# Sorting alphabetically
tasks.sort()

print("--- Current Task List ---")
print(tasks)
print(f"Recently Completed: {completed_task}\n")


# --- 2. List Comprehension (The Challenge) ---
# Goal: Create a sub-list of only "Urgent" items
# Logic: [new_item for item in list if condition]

urgent_tasks = [t for t in tasks if "Urgent" in t]

print("--- Priority Alert ---")
print(f"Urgent Items Found: {urgent_tasks}")
```
