```python
# --- 1. Cleaning Raw Input ---
raw_input = "   ID_number_99 ;   "

# strip() removes leading/trailing whitespace
# lower() standardizes the casing
# replace() swaps out specific characters
clean_input = raw_input.strip().lower().replace(";", "")

print(f"Original: '{raw_input}'")
print(f"Cleaned:  '{clean_input}'")


# --- 2. Parsing a Sentence ---
# split() breaks a string into a list based on a delimiter (default is space)
sentence = "System_Status: Operational-User_Verified-Secure"
log_tags = sentence.split("-")

print(f"\nLog List: {log_tags}")
print(f"Primary Status: {log_tags[0]}")


# --- 3. Negative Indexing ---
# Use [-3:] to grab everything from the 3rd-to-last character to the end
serial_code = "TX-994-BETA-882"
suffix = serial_code[-3:]

print(f"\nFull Serial: {serial_code}")
print(f"Last Three Characters: {suffix}")
```
