```python
# --- 1. Percentage Calculation ---
# Task: Calculate a 15% discount on a product price.
original_price = 120.00
discount_rate = 0.15

discount_amount = original_price * discount_rate
final_price = original_price - discount_amount

print(f"--- Discount Report ---")
print(f"Original Price: ${original_price}")
print(f"Discount (15%): -${discount_amount:.2f}")
print(f"Final Price: ${final_price:.2f}\n")


# --- 2. Ratio Calculation ---
# Task: Determine the aspect ratio of a screen (e.g., 1920x1080).
width = 1920
height = 1080

# To simplify a ratio, we find the Greatest Common Divisor.
# For a basic script, we can show the decimal ratio:
aspect_ratio = width / height

print(f"--- Screen Specs ---")
print(f"Resolution: {width}x{height}")
print(f"Decimal Aspect Ratio: {aspect_ratio:.2f} (or 16:9)\n")


# --- 3. Floor & Modulus Calculation ---
# Task: Calculate full crates a truck can carry and leftover items.
total_items = 525
crate_capacity = 50

# // (Floor Division) gives the number of times the divisor goes into the total
full_crates = total_items // crate_capacity

# % (Modulus) gives the remainder
leftover_items = total_items % crate_capacity

print(f"--- Logistics Report ---")
print(f"Total Items: {total_items}")
print(f"Full Crates: {full_crates}")
print(f"Items Left Over: {leftover_items}")
```
